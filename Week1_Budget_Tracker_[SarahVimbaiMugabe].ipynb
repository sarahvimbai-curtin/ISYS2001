## Part 3: Building the Budget Tracker

Use AI to help you write a program that:
- Allows users to input their monthly income
- Allows users to input up to 3 expenses  
- Calculates remaining budget
- Shows a warning if budget is negative

### 💡 Pro Tip: Be specific about requirements!

```python
# TASK: Write your detailed prompt for the budget tracker
"""
My Budget Tracker Prompt:

Write a Python budget tracker program that:
- Asks the user for their monthly income
- Asks the user to enter 3 separate expenses
- Converts all money inputs to float
- Calculates total expenses
- Calculates remaining budget
- Prints the total expenses
- Prints the remaining budget
- Displays a warning message if the remaining budget is negative
- Uses simple beginner-friendly code
- Uses f-strings for output
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
# [Paste your code here]

# Ask for monthly income
monthly_income = float(input("Enter your monthly income: "))

# Ask for 3 expenses
expense1 = float(input("Enter your first expense: "))
expense2 = float(input("Enter your second expense: "))
expense3 = float(input("Enter your third expense: "))

# Calculate total expenses
total_expenses = expense1 + expense2 + expense3

# Calculate remaining budget
remaining_budget = monthly_income - total_expenses

# Print total expenses and remaining budget
print(f"\nTotal Expenses: ${total_expenses:.2f}")
print(f"Remaining Budget: ${remaining_budget:.2f}")

# Display warning if budget is negative
if remaining_budget < 0:
    print("Warning: Your budget is negative! You are spending more than you earn.")
```

### 🐛 Debugging and Refinement

```python
# REFINED VERSION - Your Working Budget Tracker
monthly_income = float(input("Enter your monthly income: "))

expense1 = float(input("Enter your first expense: "))
expense2 = float(input("Enter your second expense: "))
expense3 = float(input("Enter your third expense: "))

total_expenses = expense1 + expense2 + expense3
remaining_budget = monthly_income - total_expenses

print(f"\nTotal Expenses: ${total_expenses:.2f}")
print(f"Remaining Budget: ${remaining_budget:.2f}")

if remaining_budget < 0:
    print("Warning: Your budget is negative! You are spending more than you earn.")
```

```python
# ENHANCED BUDGET TRACKER

monthly_income = float(input("Enter your monthly income: "))

expense1 = float(input("Enter your first expense: "))
expense2 = float(input("Enter your second expense: "))
expense3 = float(input("Enter your third expense: "))

total_expenses = expense1 + expense2 + expense3
remaining_budget = monthly_income - total_expenses

# New enhancement: calculate percentage spent
percentage_spent = (total_expenses / monthly_income) * 100

print(f"\nTotal Expenses: ${total_expenses:.2f}")
print(f"Remaining Budget: ${remaining_budget:.2f}")
print(f"Percentage of income spent: {percentage_spent:.2f}%")

if remaining_budget < 0:
    print("Warning: Your budget is negative! You are spending more than you earn.")
```
